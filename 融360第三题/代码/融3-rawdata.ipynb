# 数据预处理及特征工程

```python
# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
from scipy import sparse
```

```python
data1=pd.read_table("train_1")
data2=pd.read_table("train_2",header=None)
data2.columns=data1.columns
data3=pd.read_table("train_3",header=None)
data3.columns=data1.columns
data4=pd.read_table("train_4",header=None)
data4.columns=data1.columns
data5=pd.read_table("train_5",header=None)
data5.columns=data1.columns


#data1=data1[:1000]
```

```python
"""
提取时间特征
"""
import time, datetime
def ymd(d):
    a=np.zeros(d.shape[0])
    b=np.zeros(d.shape[0])
    c=np.zeros(d.shape[0])
    e=np.zeros(d.shape[0])
    for i in range(d.shape[0]):
        timeArray = time.strptime(d.loan_dt[i], "%Y-%m-%d")
        a[i]=timeArray.tm_year
        b[i]=timeArray.tm_mon
        c[i]=timeArray.tm_mday
        e[i]=timeArray.tm_wday
    d['year']=a
    d['mon']=b
    d['mday']=c
    d['wday']=e
ymd(data1)
ymd(data2)
ymd(data3)
ymd(data4)
ymd(data5)
#time.strptime(data1.loan_dt[0], "%Y-%m-%d")
```

```python
"""
2017-4-1的基时间戳
"""
import time, datetime
# 转为时间数组
#print("  ",str(data1.loan_dt[1]))
#type(data1.loan_dt)
#data1.loan_dt.shape
# 字符类型的时间
tt = '2017-04-01'
# 转为时间数组
timeArray0 = time.strptime(tt, "%Y-%m-%d")
print(timeArray0)     
timeStamp0 = int(time.mktime(timeArray0))//86400
print("  ",timeStamp0)  # 1381419600
```

```python
"""
时间特征
"""
def t(d):
    timeStamp=np.zeros(d.shape[0])
    for i in range(d.shape[0]):
        timeArray = time.strptime(d[i], "%Y-%m-%d")
        #print(timeArray)     
        # 转为时间戳
        timeStamp[i] = (int(time.mktime(timeArray))//86400)
        #print (timeStamp)  # 1381419600
    #print(timeStamp)
    return timeStamp-17256.0
data1.loan_dt=t(data1.loan_dt)
data2.loan_dt=t(data2.loan_dt)
data3.loan_dt=t(data3.loan_dt)
data4.loan_dt=t(data4.loan_dt)
data5.loan_dt=t(data5.loan_dt)
```

```python
data1.tag=(data1.tag=='fq')*1
data2.tag=(data2.tag=='fq')*1
data3.tag=(data3.tag=='fq')*1
data4.tag=(data4.tag=='fq')*1
data5.tag=(data5.tag=='fq')*1
```

```python
data=pd.concat([data1,data2,data3,data4,data5],axis=0)
del(data1,data2,data3,data4,data5)
data.drop(["id"],axis=1,inplace=True)
print(data[:10])
```

```python
"""
缺失值
"""
def queshi(d):
    d['queshi']=np.isnan(d[:]).sum(axis=1)
queshi(data)
```

# xgboost模型

```python

import xgboost as xgb
from sklearn.cross_validation import train_test_split
params={'booster':'gbtree',
    'objective': 'binary:logistic',
    'eval_metric':'auc',
    'gamma':0.1,
    'min_child_weight':1.1,
    'max_depth':7,
    'lambda':10,
    'subsample':0.7,
    'colsample_bytree':0.7,
    'colsample_bylevel':0.7,
    'eta': 0.01,
    'tree_method':'exact',
    'seed':1000,
    'nthread':12
    }
plst = list(params.items())

def mxgboost(train_x,train_y,test_x,test_y,plst):
    xgb_train = xgb.DMatrix(train_x, label=train_y)
    xgb_test = xgb.DMatrix(test_x, label=test_y)
    del(train_x,train_y,test_x,test_y)
    watchlist = [(xgb_train, 'train'),(xgb_test, 'val')]
    model=xgb.train(plst, xgb_train,num_boost_round=3000,evals=watchlist,early_stopping_rounds=10)
    return model
def mpre_xgb(model,data):
    xgb_data=xgb.DMatrix(data)
    return model.predict(xgb_data)
#model.save_model('xgb.model')
```

```python
model=mxgboost(train_x,train_y,test_x,test_y,plst)
```

```python
model.save_model('xgb-1514.model')
```
